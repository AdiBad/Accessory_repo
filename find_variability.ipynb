```python
import matplotlib.pyplot as plt
from collections import Counter as ct
#Total sequences
N=25
#Frequency of most frequent character in a position: n
#Number of different amino acids at a position: k
#Variability=N*k/n
        
```

```python
def variab(file_name=r"C:\Users\Aditya\Desktop\Comparative&RegulatoryGenomics\Vera\bigAss\trees\KALIGN\muscle-gene-alignment.clw"):
    file = open(file_name).read()
    segments = file.split("\n\n")
    if "sequence" in segments[0]:
        del segments[0]

    #Total sequences
    N=25
    #Frequency of most frequent character in a position: n
    #Number of different amino acids at a position: k
    #Variability=N*k/n

    full_text={}
    varia=[]
    for i in range(60):
        full_text.update({i:""})

    for seg in segments:
        #Extract frequency for all residues in these positions
        for line in seg.split("\n"):
            try:
                if line[0] in "CTFBGDADGACDHMSMSSALNSSDLS":
                    text = line.split("      ")[1]
                    if len(text)!=0:
                        k=0
                        for character in text:
                            full_text[k]=full_text[k]+character
                            k = k+1
            except:
                continue

        #Calculate variability for these positions
        for pos in range(60):
            freqs = ct(full_text[pos])
            if freqs.most_common()[0][0]=="-":
                n=freqs.most_common()[1][1]
            else:
                n=freqs.most_common()[0][1]
            if "-" in freqs:
                k = len(freqs) - 1
            else:
                k = len(freqs)
            varia.append(N*k/n)
    return (varia) 
        
    
```

```python
var = variab(r"C:\Users\Aditya\Desktop\Comparative&RegulatoryGenomics\Vera\bigAss\trees\KALIGN\muscle-species-alignment.clw")
```

```python
#DISTRIBUTION FOR 25 SPECIES
plt.scatter(range(len(var)),var,marker='.')
plt.xlabel('Sequence Length')
plt.ylabel('Variability')
plt.title('Residue distribution')
```

```python
#DISTRIBUTION FOR SELECTED GENE
v = variab(r"C:\Users\Aditya\Desktop\Comparative&RegulatoryGenomics\Vera\bigAss\trees\MUSCLE\muscle-gene-alignment.clw")
plt.scatter(range(1080),v,marker='.')
plt.xlabel('Sequence Length')
plt.ylabel('Variability')
plt.title('Residue distribution')
```
